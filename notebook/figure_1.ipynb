```python

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pathlib
import sys
sys.path.append("./../")

from src.utils import load_data

import seaborn as sns
#plt.style.use('seaborn-whitegrid')
sns.set_palette('colorblind')
sns.set_style('whitegrid')
mpl.rcParams['font.family'] = 'Arial'
```

```python
# Load data 
data = load_data([0,176])
sample_size = 3000 # TEST SIZE, if you want real plots s=3000
```

```python
# Figure 1 
sub_data = data[(data["resolution"] == 2) | (data["resolution"] == 8)]
sub_data = sub_data[(data["melt_average"] >12) & (sub_data["melt_average"] < 20)]
sub_data["resolution"] = sub_data["resolution"].astype(str) + "km"

sns.lineplot(data=sub_data,x="years",y="SLC", units="index",lw=1, hue="resolution", estimator=None, palette=sns.color_palette("colorblind")[:2])

plt.xlabel("Time (years)",fontsize=18)    
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("SLC (mm)",fontsize=18)


plt.legend(loc='upper left',fontsize=14)
plt.savefig(f"./../assets/plots/fig_1/plot.pdf",bbox_inches='tight')
```
