```python
# Data manipulation
import numpy as np
import pandas as pd

# System and paths
import pathlib

# Custom modules
from src.utils import load_data

# Plotting
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

sns.set_palette('colorblind')
sns.set_style('whitegrid')
mpl.rcParams['font.family'] = 'Arial'
```

```python
# Load data
data = load_data([0,151])
```

```python
# Figure 1
sub_data = data[(data["resolution"] == 2) | (data["resolution"] == 8)]
sub_data = sub_data[(data["melt_average"] >20) & (sub_data["melt_average"] < 30)]
sub_data["resolution"] = sub_data["resolution"].astype(str) + "km"

sns.lineplot(data=sub_data,x="years",y="SLC", units="index",lw=1, hue="resolution", estimator=None, palette=sns.color_palette("colorblind")[:2])

plt.xlabel("Time (years)",fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("SLC (mm)",fontsize=18)
plt.title("b", fontsize=18)


plt.legend(loc='upper left',fontsize=14)
plt.savefig(f"./../assets/plots/fig_1/fig.pdf",bbox_inches='tight')
```
