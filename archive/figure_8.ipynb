```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('colorblind')
sns.set_style('whitegrid')



# Generate data
means = [(2.2,10),(2.872,20),(3.285,30),(3.52,40)]
sigma = [0.5,1]
for s in sigma:
    for m,l in means:
        data = np.random.lognormal(mean=m, sigma=s, size=1000)

        plt.figure(figsize=(10, 9))

        # Plot using seaborn
        sns.distplot(data, hist=True, kde=True,
                    bins=30)
        plt.xlabel(f"Melt Average "+r'$M_{t_0}$'+f"={l} " + r'$ma^{-1}$', fontsize=32)
        plt.xticks(fontsize=22)
        plt.yticks(fontsize=22)
        plt.xlim(0,150)
        plt.ylim(0,0.1)
        plt.ylabel('Density', fontsize=32)
        plt.savefig(f"./../assets/plots/prior/mu_{l}_var_{s}.pdf",bbox_inches='tight')
```
